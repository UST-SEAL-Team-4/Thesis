```python
from project.dataset import Dataset
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
ds = Dataset()
raw_mri = ds.load_raw_mri(cohort_num=3)
stripped_mri = ds.load_skullstripped_mri(cohort_num=3)
```

```python
def compare_raw_and_stripped(raw_path, stripped_path, isGray=False):
    raw_img = nib.load(raw_path)
    raw_data = raw_img.get_fdata()
    
    stripped_img = nib.load(stripped_path)
    stripped_data = stripped_img.get_fdata()
    
    slice_index = raw_data.shape[2] // 2
    
    fig, axs = plt.subplots(1, 2, figsize=(16, 9))
    
    cmap = 'gray' if isGray else None
    
    axs[0].imshow(raw_data[:, :, slice_index], cmap=cmap)
    axs[0].set_title(f'Raw MRI scan (Slice {slice_index})')
    axs[0].axis('off')
     
    axs[1].imshow(stripped_data[:, :, slice_index], cmap=cmap)
    axs[1].set_title(f'Skull-stripped MRI scan (Slice {slice_index})')
    axs[1].axis('off')
    
    plt.show()
```

```python
compare_raw_and_stripped(raw_mri[0], stripped_mri[0])
```

```python
compare_raw_and_stripped(raw_mri[0], stripped_mri[0], isGray=True)
```
